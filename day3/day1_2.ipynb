```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("student.csv")
#sort by total
df=df.sort_values(by="total",ascending=False)
#top 5 students
top_students=df.head(5)
print(top_students)

plt.bar(top_students["name"],top_students["total"],color='blue')
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.title("Top 5 Students")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("student.csv")
#sort by total
df=df.sort_values(by="total",ascending=False)
#top 5 students
top_students=df.head(5)
plt.plot(top_students["name"],top_students["mark1"],color='red', marker='o', label='mark1', linestyle='--')
plt.plot(top_students["name"],top_students["mark2"],color='green', marker='o', label='mark2', linestyle='--')
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.legend()
plt.title("Marks of Top 5 Students")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("student.csv")
#select a first 5 students
top_students=df.head(5)
plt.pie(top_students["total"], labels=top_students["name"], autopct='%1.1f%%')
plt.title("Mark of 5 Students")
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=pd.read_csv("student.csv")
#select a last 5 students
last_students=df.tail(5)
#heatmap for last 5 students mark1 and mark2
sns.heatmap(last_students[["mark1","mark2"]], annot=True, cmap="YlGnBu")
plt.title("Heatmap of Mark1 and Mark2 for Last 5 Students")
plt.show()
```
