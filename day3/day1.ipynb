```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    'name':['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'total_score':[85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)
plt.bar(df['name'], df['total_score'], color='blue')
plt.title('Total Scores of Students')
plt.xlabel('Name')
plt.ylabel('Total Score')
plt.show()
```

```python
import matplotlib.pyplot as plt

overs = [1, 2, 3, 4, 5]
runs = [5, 12, 18, 26, 35]

# Line chart varaiya:
plt.plot(overs, runs, color='red', marker='o', linestyle='--')

# Chart-ku labels:
plt.title("Cricket Match Score")
plt.xlabel("Overs")
plt.ylabel("Total Runs")

plt.show()
```

```python
students = ['Vijay', 'Ajith', 'Surya']
marks = [85, 92, 78]

# h for horizontal
plt.barh(students, marks, color='orange') 
plt.show()
```

```python
import matplotlib.pyplot as plt

overs = [1, 2, 3, 4, 5]
india_score = [5, 12, 18, 26, 35]
aus_score = [8, 15, 20, 22, 30]

# Rendu plot, rendu label
plt.plot(overs, india_score, color='blue', marker='o', label='India')
plt.plot(overs, aus_score, color='orange', marker='*', label='Australia')

plt.title("India vs Australia Score")
plt.xlabel("Overs")
plt.ylabel("Runs")

# Indha function dhaan endha line yaaru nu oru box-la kaattum
plt.legend() 

plt.show()
```

```python
import matplotlib.pyplot as plt

players = ['Virat', 'Rohit', 'Gill', 'Rahul']
runs = [85, 45, 30, 40]

# Pie chart varaiya
plt.pie(runs, labels=players, autopct='%1.1f%%')

plt.title("Team India Runs Breakdown")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Data uruvakkuradhu
data = {
    'Maths': [95, 45, 80],
    'Science': [85, 50, 92],
    'English': [75, 60, 88]
}
# Index (Row names) la student perukala vekkurom
df = pd.DataFrame(data, index=['Vijay', 'Ajith', 'Surya'])

# 2. Heatmap varaiyurom 🌡️
sns.heatmap(df, annot=True, cmap='Blues')

plt.title("Student Marks Heatmap")
plt.show()
```
