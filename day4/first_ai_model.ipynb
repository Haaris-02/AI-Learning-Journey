```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print(df.head(5))
df['Sex']=df['Sex'].map({'male':0,'female':1})
df['Age']=df['Age'].fillna(df['Age'].mean())
y=df['Survived']
X=df[['Pclass', 'Sex', 'Age', 'Fare']]

from sklearn.model_selection import train_test_split
# 1. Decision Tree Algorithm-a ulla kondu varuvadhu
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


# 2. Oru pudhu, empty aana AI model-a uruvakkuradhu
model = DecisionTreeClassifier()

# 3. Model-a Train pandradhu (Idhudhaan Magic step! ✨)
model.fit(X_train, y_train)

print("AI Model Training Mudinjiduchu! 🚀")

score = model.score(X_test, y_test)
print(f"AI Model Score: {score}")

import numpy as np

# Jack matrum Rose details-a oru list-kulla list aaga tharuvom
new_passengers = [
    [3, 0, 20, 8],   # Jack
    [1, 1, 17, 71]   # Rose
]

# AI kitta predict panna sollurom
predictions = model.predict(new_passengers)

print(f"Jack's Survival: {predictions[0]} (0=Died, 1=Survived)")
print(f"Rose's Survival: {predictions[1]} (0=Died, 1=Survived)")
```
