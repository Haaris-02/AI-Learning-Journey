```python
import pandas as pd
data = {
    'movie': ['Inception', 'The Matrix', 'Interstellar'],
    'rating': [8.8, 8.7, 8.6]
}
df = pd.DataFrame(data)
print(df[df['rating'] > 8.7])
```

```python
#data filtering in pandas
import pandas as pd
data = {
    'movie': ['Inception', 'The Matrix', 'Interstellar'],
    'rating': [8.8, 8.7, 8.6]
}
df = pd.DataFrame(data)
print(df[df['rating'] > 8.7])
print(df[df['movie'] == 'Inception'])
```

```python
#data cleaning in pandas
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', None, 'Charlie'],
    'Age': [25, None, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago', None]
}
df = pd.DataFrame(data)
print("Before cleaning:")   
print(df)
df_cleaned = df.dropna()
print("\nAfter cleaning:")
print(df_cleaned)   
print("\nFilling missing values:")
df_filled = df.fillna({'Name': 'Unknown', 'Age': df['Age'].mean(), 'City': 'Unknown'})
print(df_filled)    
```

```python
#csv file handling in pandas
import pandas as pd
# Reading a CSV file
df = pd.read_csv('student.csv')
print("Data from CSV file:")
print(df)
# Writing to a CSV file
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
df.to_csv('output.csv', index=False)
print("Data written to output.csv")

```

```python
#csv file student.csv add mark total column
import pandas as pd
# Reading the CSV file
df = pd.read_csv('student.csv')
#add total column
df['Total'] = df['mark1'] + df['mark2']
df['Percentage'] = df['Total'] / 200 * 100
print(df)
failed_students = df[df['mark1'] < 70]
print("\nFailed students:")
print(failed_students)
#mark1 and mark2 average
average_mark1 = df['mark1'].mean()
average_mark2 = df['mark2'].mean()
print(f"\nAverage mark1: {average_mark1}")
print(f"Average mark2: {average_mark2}")
#print max total student name
print(df['name'][df['Total'].idxmax()])
```
