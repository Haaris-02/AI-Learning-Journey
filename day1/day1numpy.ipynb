```python
#indexing and slicing in 2D and 3D array
import numpy as np
a1=np.arange(24).reshape(6,4)
a2=np.arange(24).reshape(2,3,4)
#2D array indexing and slicing
print('Element at row 1, column 2:', a1[1, 2])  # element at row 1, column 2
print('Row 1:', a1[1])  # row 1
print('Column 2:', a1[:, 2])  # column 2
```

```python
#3D array indexing and slicing
print('3D Array a2:\n', a2)
print('Element at block 0, row 1, column 2:', a2[0, 1, 2])  # element at block 0, row 1, column 2
print('Block 0:', a2[0])  # block 0
print('Row 1 across all blocks:', a2[:, 1, :])  # row 1 across all blocks
print('Column 2 across all blocks:', a2[:, :, 2])  # column 2 across all blocks
print('Slicing block 0, rows 0-1, columns 1-3:', a2[0, 0:2, 1:4])  # slicing block 0, rows 0-1, columns 1-3
```

```python
#aggregation /statistics in 1D 2D and 3D array
import numpy as np
a1=np.arange(24)
a2=np.arange(24).reshape(6,4)
a3=np.arange(24).reshape(2,3,4)
#Aggregation in 1D array
print('1D Array a1:\n', a1)
print('Sum:', np.sum(a1))
print('Mean:', np.mean(a1))
print('Standard Deviation:', np.std(a1))
print('Variance:', np.var(a1))
#Aggregation in 2D array
print('2D Array a2:\n', a2)
print('Sum:', np.sum(a2))
print('Mean:', np.mean(a2))
print('Standard Deviation:', np.std(a2))
print('Variance:', np.var(a2))
#Aggregation in 3D array
print('3D Array a3:\n', a3)
print('Sum:', np.sum(a3))
print('Mean:', np.mean(a3))
print('Standard Deviation:', np.std(a3))
print('Variance:', np.var(a3))
```
